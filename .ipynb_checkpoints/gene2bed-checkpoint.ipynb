```python
s = df_gtf.loc[0, 8]
s
```

```python
def get_gene_name(s):
    l = [x.split() for x in s.split(';')[:-1]]
    return [x[1].replace('"', '') for x in l if x[0] == 'gene_name'][0]
```

```python
get_gene_name(s)
```

```python
%%time
df_gtf['gene_name'] = df_gtf.loc[:,8].map(get_gene_name)
```
